```julia
using MathematicalSystems
using InvariantSets
using Plots
```

```julia
A = [0.9 0.5; 0 0.9]
B = [1., 0]
D = I(2)
X = BallInf(zeros(2), 10.)
U = BallInf(zeros(1), 1.)
W = BallInf(zeros(2), 0.2)

K = [0.1 0.1]

autSys = @system x⁺ = A*x  x∈X
ctrlSys = @system x⁺ = A*x + B*u x∈X u∈U
noisyCtrlSys = @system x⁺ = A*x + B*u + D*w x∈X u∈U w∈W;
```

```julia
Xfeas = state_constraints(ctrlSys, K)
plot(X)
plot!(convert(HPolytope, Xfeas))
```

```julia
IS = maximum_invariant_set(autSys)
plot(X)
plot!(IS)
```

```julia
using CDDLib
using Polyhedra

MIL = maximum_invariant_set(ctrlSys, K)
MCI = maximum_control_invariant_set(ctrlSys)
plot(X)
plot!(MCI)
plot!(MIL)
```

```julia
using CDDLib
using Polyhedra

preSetaut = preset(autSys, X)
preSet = preset(ctrlSys, X)
Xf = feasible_set(ctrlSys, X, 20)
plot(X)
plot!(preSetaut)
plot!(preSet)
plot!(convert(VPolytope, Xf))
plot!()
```
