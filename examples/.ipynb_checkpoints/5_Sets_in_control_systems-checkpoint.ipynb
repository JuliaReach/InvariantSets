```julia
# Load Dependencies
using Plots
using InvariantSets
using LinearAlgebra
```

```julia
A = [1. 1; 0 1]
B = Matrix([0.5 1]')
X = HPolyhedron([0. 1.], [2.])
U = BallInf(zeros(1), 1.)
W = BallInf(zeros(2), 0.1)
D = Diagonal(ones(2))
Q = I
R = 0.01*I
K = [0.660853  1.32606]
P = [1.63408  -0.625;
     -0.625     1.61612]
# P = [2.0066 0.5099; 0.5099 1.2682]
N = 9
x0 = [-5,  -2.]
x0 = [-5.2,  -2.]
S = A - B*K;
```

### Terminal Set

```julia
# Ω = minimal_RPI_set(InvariantSets.Rakovic(), S, W)
# X_bar = X -ᶜ Ω
# U_bar = U -ᶜ K*ᶜΩ
# Xf = terminal_set(A, B, X_bar, U_bar, K)
Xf = terminal_set(A, B, X, U, K)
plot(X, label="state constraints", alpha=0.25); # plot!(X_bar, label="tightened state constraints", alpha=0.25);
# plot!(Xf + Ω, label="terminal set");
plot!(Xf, alpha=1., color=:darkgreen, label="nominal terminal set")
```

### Feasible Sets

```julia
using Polyhedra, CDDLib

# X_feas = feasible_set(A, B, X_bar, U_bar, Xf, 9)
X_feas = feasible_set(A, B, X, U, Xf, 9)
plot(X, label="state constraints", alpha=0.25); #plot!(X_bar, label="tightened state constraints", alpha=0.25);
# plot!(X_feas + Ω, label="Feasible set for real system at N=9")
plot!(X_feas, label="Feasible set for nominal system at N=9", alpha=1., color=:darkgreen)
plot!(Xf)
```
